```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\feras\\JupyterNoteBook\\LinearRegresstion\\homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel("Area(sqr ft)")
plt.ylabel("Price(US$)")
plt.scatter(df.area,df.price)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df[['price']])
```

```python
reg.predict([[3300]]) # if we want to predict this number
```

```python
reg.predict([[6060]]) # if we want to predict this number
```

```python
reg.coef_ # m - slope
```

```python
reg.intercept_ # b - Intercept
```

```python
#The prediction comes from the previous equation 
# y = m*x+b
y = 135.78767123 * (3300) + 180616.43835616432
y
```

```python
# Let's try on diffrent areas (Array of areas)
area = pd.read_csv('Areas.csv')
area
```

```python
predected_areas = reg.predict(area)
predected_areas
```

```python
area['price'] = predected_areas # Add new coulmn
```
