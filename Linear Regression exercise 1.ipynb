```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\feras\\JupyterNoteBook\\LinearRegresstion\\anada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.year,df[['income']],color='red')
plt.plot(df.year,df[['income']],color='blue')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df[['income']])
```

```python
reg.predict([[2020]]) # Predict this year income 2020
```
